```python
energy_kcal = [-13.4, -2.7, 5.4, 42.1]
print(energy_kcal)
```

```python
for number in energy_kcal:
    kJ = number*4.184
    print(kJ)
```

```python
energy_kJ = []
for number in energy_kcal:
    kJ = number*4.184
    energy_kJ.append(kJ)
print(energy_kJ)
```

```python
negatives = []
positives = []

for number in energy_kcal:
    if number<0:
        negatives.append(number)
    else:
        positives.append(number)
```

```python
print(negatives, '\n', positives)
```
